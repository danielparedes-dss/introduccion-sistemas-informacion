#Conclusiones
A partir del análisis del dataset del Titanic, se identificaron varios aspectos relevantes. El porcentaje de supervivencia fue del 38.38%, lo que indica que la mayoría de los pasajeros no logró sobrevivir al accidente. Este resultado sugiere que se trató de un evento de gran magnitud, en el que las condiciones de respuesta ante la emergencia pudieron no haber sido suficientes.

Esto indica que una gran parte de los viajeros correspondía a personas de recursos limitados. Considerando el contexto histórico, es posible que muchos de estos pasajeros fueran migrantes que se dirigían a Estados Unidos en busca de mejores oportunidades.

Respecto al género, se evidenció que había una mayor cantidad de hombres que de mujeres a bordo. Este comportamiento podría estar relacionado con dinámicas sociales de la época.
